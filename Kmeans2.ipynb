```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans
```

```python
#using the make_blobs function from the sklearn.datasets module we generate the data 
#The centers parameter represents the number of clusters.
#cluster_std The standard deviation of the clusters.
```

```python
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.55, random_state=0)
plt.scatter(X[:,0], X[:,1])
```

```python
#the optimal number of clusters is determined using the elbow method
#we train multiple models using a different number of clusters 
#and we store the value of the intertia_ property (Within Cluster Sum of Squares (WCSS)) every time.

#class sklearn.cluster.KMeans(...), n_clusters the number of clusters to form as well as the number of centroids to generate.
#init Method for initialization, ‘k-means++’ : selects initial cluster centers in a smart way to speed upconvergence. 
#n_init Number of time the k-means algorithm will be run with different centroid seeds. 
#max_iter Maximum number of iterations of the k-means algorithm for a single run.
```

```python
wcss = []
for i in range(1, 21):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 21), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#using the optimum number of clusters (4) we categorize data using K means
#kmeans.cluster_centers_[] coordinates of cluster centers. 
```

```python
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = kmeans.fit_predict(X)
plt.scatter(X[:,0], X[:,1])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.show()
```
